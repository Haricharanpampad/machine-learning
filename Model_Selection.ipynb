```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
heart_data=pd.read_csv('/content/heart.csv')
```

```python
heart_data.head()
```

```python
X=heart_data.drop(columns='target',axis=1)
Y=heart_data['target']
```

```python
X=np.asarray(X)
Y=np.asarray(Y)
```

#Model selection
comparing models with default hyper parameters using cross validation

```python
models=[LogisticRegression(max_iter=1000),SVC(kernel='linear'),KNeighborsClassifier(),RandomForestClassifier(random_state=0)]
```

```python
def compare_models_cross_validation():
  for model in models:
    cv_score=cross_val_score(model,X,Y,cv=5)
    mean_accuracy=sum(cv_score)/len(cv_score)
    mean_accuracy=mean_accuracy*100
    mean_accuracy=round(mean_accuracy,2)
    print('cross val accuracies for model',model,'=',cv_score)
    print('accuracy score of ',model,'=',mean_accuracy,'%')
    print('------------------------------------')

```

```python
compare_models_cross_validation()
```

Raandom forest  classifier as highest accuracy

comparing models with  hyper tuned  parameters using cross validation
technique(Grid search CV)

```python
model_list=[LogisticRegression(max_iter=10000),SVC(),KNeighborsClassifier(),RandomForestClassifier(random_state=0)]
```

```python
#dictionary of hyperparameters
model_hyperparameters={
    'log_reg_hyperparameters':{
        'C':[1,5,10,20]
    },
    'svc_hyperparameters':{
        'kernel':['linear','poly','rbf','sigmoid'],
        'C':[1,5,10,20]
    },
    'KNN_hyperparameters':{
        'n_neighbors':[3,5,10]
    },
    'random_forest_hyperparameters':{
        'n_estimators':[10,20,50,100]
    }
}
```

```python
type(model_hyperparameters)
```

```python
model_keys=list(model_hyperparameters.keys())
print(model_keys)
```

Applying GridSearchCV

```python
def ModelSelection(list_of_models,hyperparameters_dic):
  result=[]
  i=0
  for model in list_of_models:
    key=model_keys[i]
    params=hyperparameters_dic[key]
    i+=1
    print(model)
    print(params)
    print('--------------------------------------------------')

    classifier=GridSearchCV(model,params,cv=5)
    classifier.fit(X,Y)

    result.append({
        'model used':model,
        'highest score':classifier.best_score_,
        'best hyperparams':classifier.best_params_
    })
  result_dataframe=pd.DataFrame(result,columns=['model used','highest score','best hyperparams'])
  return result_dataframe
```

```python
ModelSelection(model_list,model_hyperparameters)
```
